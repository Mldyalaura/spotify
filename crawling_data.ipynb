import library

```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
#scrape all available reviews
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.spotify.music',
    lang = 'id',
    country = 'id',
    sort = Sort.NEWEST,
    count=10000,
    filter_score_with=None
)
```

show data yang sudah di crawling

```python
#df_busu = busu itu variabel jadi bisa di ketik apa saja
df_busu = pd.DataFrame(np.array(result),columns=['review'])
df_busu = df_busu.join(pd.DataFrame(df_busu.pop('review').tolist()))
df_busu.tail()
```

ambil data yang dibutuhkan

```python
df_busu[['userName', 'score', 'at', 'content']].tail()
```

```python
my_df = df_busu[['userName', 'score', 'at', 'content']]
```

```python
#simpan dataset ke dalam file
my_df.to_csv("DataSetSpotify.csv", index = False)
```
